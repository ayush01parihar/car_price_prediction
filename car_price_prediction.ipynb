# handling missing values

```python
#1 replacing ? with nan
df.replace("?",np.nan,inplace=True)

#2 changing dtype of column
df["normalized-losses"]=df["normalized-losses"].astype(float)
df["horsepower"]=df["horsepower"].astype(float)

#3 saving the mean of the column
nmean=df["normalized-losses"].mean()
hmean=df["horsepower"].mean()

#4 fill nan values with mean
df["normalized-losses"].fillna(nmean,inplace=True)
df["horsepower"].fillna(hmean,inplace=True)
```

# Model Creation

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.35,random_state=1)

from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(xtrain,ytrain)
ypred = linreg.predict(xtest)

from sklearn.metrics import r2_score
print(r2_score(ytest,ypred))
```

# Bias and Variance

```python
train = linreg.score(xtrain,ytrain)
test = linreg.score(xtest,ytest)
print(f"Training Accuracy:- {train}\n Testing Accuracy:- {test}")
```

```python
#High Training Accuracy :- Low Training Error :- Low Bias
#Low Testing Accuracy :- High Testing Error :- High Variance
#Low Bias + High Variance ---> Overfitting
```

# Ridge or Lasso

```python
from sklearn.linear_model import Ridge, Lasso

l2 = Ridge(alpha=4)
l2.fit(xtrain,ytrain)
ypred = l2.predict(xtest)
```

```python
train = l2.score(xtrain,ytrain)
test = l2.score(xtest,ytest)
print(f"Training Accuracy:- {train}\n Testing Accuracy:- {test}")
```

# Hyperparameter Tuning

```python
for i in range(1,50):
    l2 = Ridge(alpha=i)
    l2.fit(xtrain,ytrain)

    train = l2.score(xtrain,ytrain)
    test = l2.score(xtest,ytest)

    print(f"{i} {train} {test} {train - test}")
```

```python
l2 = Ridge(alpha=3)
l2.fit(xtrain,ytrain)
ypred = l2.predict(xtest)
```
